```python
# Dependencies and Setup
# %matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, on = 'Mouse ID', how = 'outer')
# Display the data table for preview
print(study_results.head())
print(mouse_metadata.head())

```

```python
# Checking the number of mice.
len(df['Mouse ID'].value_counts())
```

```python
duplicate_ids_test = study_results.drop_duplicates
duplicate_ids_test
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# 1.) create a 

# ms_clnd_study = study_results.loc[:, ['Mouse ID', 'Timepoint']]
# print(ms_clnd_study)

#tells num of dup rows
duplicate_ids = df.loc[df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

print(duplicate_ids)
id_duplicate_data = df.loc[df['Mouse ID'] == duplicate_ids[0]]
id_duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = df[df['Mouse ID'].isin(duplicate_ids)==False]
clean_df.head(40)
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df['Mouse ID'].value_counts())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Assemble the resulting series into a single summary dataframe.
# 1.) calculate the numbers for each of these 2.) populate a dataframe with the values

# mean, median, std dev, SEM, var
tmr_vol_mean = clean_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
tmr_vol_median = clean_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
tmr_vol_std_dev = clean_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
tmr_vol_SEM  = clean_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

# # what's the code for variance?
tmr_vol_var = clean_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']

# populate dataframe
sum_stats = {'Mean': tmr_vol_mean,
            'Median': tmr_vol_median,
            'Variance': tmr_vol_var,
            'Std Dev': tmr_vol_std_dev,
            'SEM': tmr_vol_SEM}
sum_stats_df = pd.DataFrame(sum_stats)
sum_stats_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# see above
tmr_vol_df = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)']
# Using the aggregation method, produce the same summary statistics in a single line
tmr_vol_agg_df = tmr_vol_df.agg(['mean','median','var', 'std', 'sem'])
tmr_vol_agg_df

```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.
# first create a data frame out of the data
uniq_mice_variable = len(clean_df['Mouse ID'].value_counts())
tot_uniq_mice_df = pd.DataFrame([{'Unique Mice': uniq_mice_variable}])
tot_uniq_mice_df

# generate the bar chart from this
mice_bar = tot_uniq_mice_df.plot(kind = 'bar')
title = 'Number of unique mice tested = ' + str(uniq_mice_variable)
plt.title(title)
mice_bar.set_xticklabels('Unique Mice', rotation = 0)

plt.show()
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.
uniq_mice_per_drug = clean_df.groupby('Drug Regimen').count()[['Mouse ID']]
u_m_p_d = uniq_mice_per_drug.reset_index()
md_xaxis = u_m_p_d['Drug Regimen'].to_list()
md_yaxis = u_m_p_d['Mouse ID'].to_list()
plt.bar(md_xaxis,md_yaxis)
plt.xticks(rotation=90)
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
male_fem_df = clean_df.groupby('Sex')[['Mouse ID']].count()
m_f_pie = male_fem_df.plot(kind = 'pie', subplots = True)
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
male_fem = clean_df.groupby('Sex').count()[['Mouse ID']].reset_index()
m_f_dist = male_fem['Mouse ID'].to_list()
plt.pie(m_f_dist)
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
mouse_final_df_Cap = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']
mouse_final_df_Ram = clean_df.loc[clean_df['Drug Regimen'] == 'Ramicane']
mouse_final_df_Inf = clean_df.loc[clean_df['Drug Regimen'] == 'Infubinol']
mouse_final_df_Cef = clean_df.loc[clean_df['Drug Regimen'] == 'Ceftamin']
mouse_final_df_four = pd.concat([mouse_final_df_Cap, mouse_final_df_Ram, mouse_final_df_Inf, mouse_final_df_Cef], axis = 0)
mouse_final_df_four

# Start by getting the last (greatest) timepoint for each mouse
# I found this soln on stack overflow.  Originally I was going to create a for and while loop
# to go through and find the last item in each row but instead found the .ne function
# this not only find the lasts item, but it also allows me to create a dataframe more cleanly
# to display the final data desired

#identify the final Timepoint for each mouse
mouse_final_df_four['Changed'] = mouse_final_df_four['Mouse ID'].ne(mouse_final_df_four['Mouse ID'] \
                                                                        .shift(-1).bfill()).astype(int)

# filter to give the final tumor volume
mouse_final_tumor = mouse_final_df_four.loc[mouse_final_df_four['Changed'] == 1]
mouse_final_tumor.head()
```
